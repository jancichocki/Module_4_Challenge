```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Define the tickers for 2-year and 10-year Treasury bonds
# These tickers can vary based on the data provider
ticker_2y = "^FVX"  # Example ticker for 2-year Treasury bond yield
ticker_10y = "^TNX"  # Example ticker for 10-year Treasury bond yield

# Fetch historical data
data_2y = yf.download(ticker_2y, start="2020-01-01", end="2021-01-01")['Close']
data_10y = yf.download(ticker_10y, start="2020-01-01", end="2021-01-01")['Close']

# Resample the data to get monthly average yields
monthly_data_2y = data_2y.resample('M').mean()
monthly_data_10y = data_10y.resample('M').mean()

# Create a DataFrame for plotting
df = pd.DataFrame({
    '2 Year': monthly_data_2y,
    '10 Year': monthly_data_10y
})

# Plotting
df.plot(figsize=(10, 6), title='2-Year vs 10-Year US Treasury Bond Yields')
plt.ylabel('Yield (%)')
plt.xlabel('Date')
plt.show()
```
