```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas  # Assuming you have hvplot installed; if not, you might use matplotlib or seaborn
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    Path("Resources/crypto_market_data.csv"),
    index_col="coin_id")

# Display sample data
print(df_market_data.head(10))

# Generate summary statistics
print(df_market_data.describe())

# Plot your data to see what's in your DataFrame
# Note: Adjust the plotting library if you're not using hvplot
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)



# Normalize the data
scaled_data = StandardScaler().fit_transform(df_market_data)

# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(
    scaled_data,
    columns=df_market_data.columns,
    index=df_market_data.index)

# Display sample data
print(df_market_data_scaled.head())





# Set style for seaborn
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

# Scatter plot for PC1 vs PC2 colored by Cluster_PCA
sns.scatterplot(x="PC1", y="PC2", hue="Cluster_PCA", palette="viridis", data=df_pca, s=100, alpha=0.7).set_title("PCA: PC1 vs PC2 Clustering")
plt.legend(title='Cluster')
plt.show()

# Scatter plot for PC1 vs PC3 colored by Cluster_PCA
plt.figure(figsize=(12, 8))
sns.scatterplot(x="PC1", y="PC3", hue="Cluster_PCA", palette="viridis", data=df_pca, s=100, alpha=0.7).set_title("PCA: PC1 vs PC3 Clustering")
plt.legend(title='Cluster')
plt.show()

# Scatter plot for PC2 vs PC3 colored by Cluster_PCA
plt.figure(figsize=(12, 8))
sns.scatterplot(x="PC2", y="PC3", hue="Cluster_PCA", palette="viridis", data=df_pca, s=100, alpha=0.7).set_title("PCA: PC2 vs PC3 Clustering")
plt.legend(title='Cluster')
plt.show()






inertia = []
k_range = range(1, 12)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_market_data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve using matplotlib or an alternative
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()



# Assuming the best k is found (replace `4` with your best k value)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df_market_data_scaled)
df_market_data['Cluster'] = kmeans.labels_

# Example scatter plot (adjust according to your best k and insights)
# This might require adaptation if using libraries other than hvplot
df_market_data.hvplot.scatter(x="price_change_percentage_24h", y="price_change_percentage_7d", by="Cluster", hover_cols=['coin_id'])



pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_market_data_scaled)

print(f"Explained Variance: {pca.explained_variance_ratio_}")

# Turn the PCA results into a dataframe for further use
df_pca = pd.DataFrame(
    data=df_pca,
    columns=['PC1', 'PC2', 'PC3'],
    index=df_market_data.index)

# Fit K-Means using the PCA data
kmeans_pca = KMeans(n_clusters=4, random_state=42)  # Adjust k based on PCA analysis
kmeans_pca.fit(df_pca)
df_pca['Cluster_PCA'] = kmeans_pca.labels_
```
